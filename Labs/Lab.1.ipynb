Mathematical identity: 
$$ \sum_{k=1}^{n} k = \frac {n(n+1)}{2} $$

```python
## define a function that does the mathematical identity ##
def summ(n):
    num = 0            ## initial 'arbitrary number' ##
    sum_num = 0        ## initial sum ##
    for i in range (n):
        num += 1       ## increment by 1 ##
        sum_num += num ## increment by num ##
    return sum_num     ## returns the summation ##

## test the first 10 values of n starting from 0 ##
for i in range(10):
    print("summ function of n =" , i , "is " , summ(i))
```

Checking **summ(n)** equals $\frac {n(n+1)}{2} $

```python
## Test summ(n) for i = 1, 2, ... , 1000 ##
print("summ(1000) is ", summ(1000))

n = 1000
for i in range(1, n + 1):
    print("summ function of n=", i , "is ", summ(i))
```
